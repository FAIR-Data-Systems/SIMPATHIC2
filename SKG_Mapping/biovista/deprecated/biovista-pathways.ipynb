```ruby
require 'json'
require 'rest-client'
require '../OBO_Lookup/obo.rb'

ONTOLOGY = "GO".freeze
CSVFILE = "./raw_data/bv-disease-graph.large".freeze
COLUMN = 0
OUTPUT = "./maps/biovista-pathways.map".freeze
```

```ruby
out = File.open(OUTPUT, "w");

@seen = {}
File.readlines(CSVFILE, chomp: true).each do |line|
#CSV.foreach(CSVFILE) do |row|
  row = line.split("\t")
  next unless row[COLUMN + 1] == "Pathway"
  value = row[COLUMN]
  value.strip!
  next if @seen[value]
  # warn "looking for #{value}"
  if value.empty?
    warn "no value found in the column"
    next
  end
  n = NCBO.new()
  uri = n.search_by_term(term: value, ontology: ONTOLOGY, exact: true)
  unless uri  # backup plan 
    uri = n.search_by_term(term: value, ontology: "MESH", exact: true) # returns list
  end
  unless uri  # backup plan 
    uri = n.search_by_term(term: value, ontology: "SNOMEDCT", exact: true) # returns list
  end
  unless uri  # give up
    out.write  "No URI found for #{value}\n"
  else
    out.write  "#{uri},#{value}\n"
  end
  @seen[value] = 1
end
```
