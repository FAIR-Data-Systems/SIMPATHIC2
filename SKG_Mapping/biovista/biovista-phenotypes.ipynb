```ruby
require 'json'
require 'rest-client'
require '../OBO_Lookup/obo.rb'

ONTOLOGY = "HP".freeze
CSVFILE = "./raw_data/bv-disease-graph.large".freeze
COLUMN = 0
OUTPUT = "./maps/biovista-phenotypes.map".freeze
```

```ruby
out = File.open(OUTPUT, "w");

@seen = {}

File.readlines(CSVFILE, chomp: true).each do |line|
  row = line.split("\t")
  next unless row[COLUMN + 1] == "Human Phenotype"
  value = row[COLUMN]
  value.strip!
  next if @seen[value]
  warn "looking for #{value}"
  if value.empty?
    warn "no value found in the column"
    next
  end
  n = NCBO.new()
  uri = n.search_by_term(term: value, ontology: ONTOLOGY, exact: true)
  unless uri  # backup plan NDFRT
    uri = n.search_by_term(term: value, ontology: "EFO", exact: true) # returns list
  end
  unless uri  # give up
    out.write "No URI found for #{value}\n"
  else
    out.write "#{uri},#{value}\n"
  end
  @seen[value] = 1
end
out.close

puts "done"
```
