```ruby
require 'net/http'
require 'uri'
require 'json'

# List of MeSH CUIs
mesh_cuis = %w[C0268631 C0751882 C0023264 C0043459 C0268467 C0349653 C0027126 C1849508 C0024408 C0268595]
api_key = "24e04058-54e0-11e0-9d7b-005056aa3316"

mesh_cuis.each do |cui|
  # Construct the BioPortal mappings URL
  # Note: MeSH CUIs may need mapping to descriptor IDs (DXXXXXX) or SCR IDs (CXXXXXX)
  # Here, we assume CUIs are valid as SCRs; adjust if descriptors are needed
  class_uri = "http://purl.bioontology.org/ontology/MESH/#{cui}"
  encoded_uri = URI.encode_www_form_component(class_uri)
  url = "https://data.bioontology.org/ontologies/MESH/classes/#{encoded_uri}/mappings?ontologies=LOOM&apikey=#{api_key}"

  # Parse the URL
  uri = URI.parse(url)

  # Make the HTTP GET request
  http = Net::HTTP.new(uri.host, uri.port)
  http.use_ssl = true
  response = http.get(uri.request_uri)

  # Check response
  if response.is_a?(Net::HTTPSuccess)
    data = JSON.parse(response.body)
    found = false
    data.each do |mapping|
      mapping['classes'].each do |cls|
        if cls['@id']&.start_with?('http://www.orpha.net/ORDO/')
          orpha_id = cls['@id'].split('/').last
          puts "MeSH CUI: #{cui}, ORPHAcode: #{orpha_id}"
          found = true
        end
      end
    end
    puts "MeSH CUI: #{cui}, No ORPHAcode found" unless found
  else
    puts "Error for MeSH CUI #{cui}: HTTP #{response.code} - #{response.message}"
  end
end
```
