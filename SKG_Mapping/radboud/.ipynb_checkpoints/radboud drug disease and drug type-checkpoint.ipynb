```ruby
require_relative '../Lookups/metadata_functions'
require 'csv'

#f1 = File.open('./mappings/drug-drugtype-mappings.csv', 'w')
f2 = File.open('./mappings/drug-disease-mappings.csv', 'w')
e = File.open('./mappings/drug-disease-drugtype-errors.txt', 'w')
#f1.sync = true # Ensure immediate writes
f2.sync = true # Ensure immediate writes
e.sync = true # Ensure immediate writes
#f1.write "radboudsource,sourceguid,sourcelabel,targettype\n"
f2.write "radboudsource,sourceguid,sourcelabel,radboudtarget,targetguid,targetlabel\n"


# CSV.foreach('./rawdata/drug-drugtype.csv', headers: true) do |row|
  
#   next if row.size < 3

#   radbouddrug = row[0]
#   # always CHEMBL
#   lookup = CHEMBL.new(uri: radbouddrug)

#   drugurl = "http://rdf.ebi.ac.uk/resource/chembl/molecule/#{radbouddrug}"

#   drugtitle, moleculetype = lookup.lookup_title_and_type
#   # lookup.lookup_title_by_uri(term_uri: url, ontology: ontology)
#   if drugtitle
#     warn "#{radbouddrug}, #{drugurl}, #{drugtitle}"
#   else # this shoujld never happen because the raw curie is returned if it fails
#     warn "No match for #{radbouddrug}, #{drugurl}"
#     e.write "No match for #{radbouddrug}, #{drugurl}"
#     next
#   end
#   f1.write CSV.generate_line([radbouddrug, drugurl, drugtitle, moleculetype])
# end

CSV.foreach('./rawdata/drug-disease.csv', headers: true) do |row|
  radbouddrug = row[0]
  # always CHEMBL
  lookup = CHEMBL.new(uri: radbouddrug)

  drugurl = "http://rdf.ebi.ac.uk/resource/chembl/molecule/#{radbouddrug}"

  drugtitle, moleculetype = lookup.lookup_title_and_type
  # lookup.lookup_title_by_uri(term_uri: url, ontology: ontology)
  if drugtitle
    warn "#{radbouddrug}, #{drugurl}, #{drugtitle}"
  else # this shoujld never happen because the raw curie is returned if it fails
    warn "No match for #{radbouddrug}, #{drugurl}"
    e.write "No match for #{radbouddrug}, #{drugurl}"
    next
  end

  radbouddisease = row[2]
  ontology = radbouddisease.match(/([^_]+)_/)[1]

  if ontology == "EFO"
    diseaseurl = "http://www.ebi.ac.uk/efo/#{radbouddisease}"
  elsif ontology == "OTAR"
    diseaseurl = "http://www.ebi.ac.uk/efo/#{radbouddisease}"
  elsif ontology == "DOID"
    diseaseurl = "https://api.disease-ontology.org/v1/terms/#{radbouddisease}"
  elsif ontology == "Orphanet"
    diseaseurl = "http://www.orpha.net/ORDO/#{radbouddisease}"
  else
    diseaseurl = "http://purl.obolibrary.org/obo/#{radbouddisease}"
  end

  diseasetitle = ontology_annotations(uri: diseaseurl)
  # lookup.lookup_title_by_uri(term_uri: url, ontology: ontology)
  if diseasetitle
    warn "#{radbouddisease}, #{diseaseurl}, #{diseasetitle}"
  else # this shoujld never happen because the raw curie is returned if it fails
    warn "No match for #{radbouddisease}, #{diseaseurl}"
    e.write "No match for #{radbouddisease}, #{targediseaseurlturl}"
    next  
  end

  f1.write CSV.generate_line([radboudsdrug, drugurl, drugitle, radbouddisease, diseaseurl, diseasetitle])

end

f1.close
f2.close
e.close
```
