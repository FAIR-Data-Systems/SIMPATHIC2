```ruby
require 'csv'

# puts `pwd`
# puts `ls -l`
gene = {}
CSV.foreach('./mappings/drug-gene-mappings.csv', headers: true) do |row|
  gene[row[2]] = row[3] # 2 is the ENS-ID, 3 is the UniProt mapping
end

puts gene.first


CSV.foreach('./rawdata/drug-reactome.csv', headers: true) do |row| # NOT DRUG - it is genes, actually
  ensid = row[0]
  abort "not found #{row[0]}" unless gene[ensid]
end
puts "all good!"  # we already know all of these genes
```

# Process file

using the existing mapping files rather than remote lookups



```ruby
require_relative '../Lookups/metadata_functions'
require 'csv'

f = File.open('./mappings/gene-reactome.csv', 'w')
e = File.open('./mappings/gene-reactome-errors.txt', 'w')
f.sync = true # Ensure immediate writes
e.sync = true # Ensure immediate writes
f.write "radboudsource,sourceguid,target,targetguid,targetlabel\n"

CSV.foreach('./rawdata/drug-reactome.csv', headers: true) do |row|  # not drug, gene!
  
  next if row.size < 3
  radboudsource = row[0]
  sourceguid = gene[radboudsource]
  target = row[2]
  targetguid = "http://identifiers.org/reactome/#{target}"    # this is the URI used by UniProt
  r = REACTOME.new(uri: target)
  targetlabel = r.lookup_title
  warn targetlabel

  f.write CSV.generate_line([radboudsource,sourceguid,target,targetguid,targetlabel])

end

f.close
e.close
puts "done"
```
